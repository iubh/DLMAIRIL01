```python
### Import required modules.

import matplotlib.pyplot as plt 
import math
from array import array
```

```python
### This will be V(x,a). We only required a fixed amount of memory. 

valueActionFunction = array('l')
valueActionFunction = [0,1,2,3,4,5]
```

```python
### This will be the value function V(x). We will allocate memory dynamically. 

valueFunction = array('l')
valueFunction = []
```

```python
### This will be the policy function phi(x). We will allocate memory 
### dynamically. 

policyFunction = array('i')
policyFunction = []
```

```python
### We need temporary storage for the current value of V(x) witin our 
### iterations. The initial value MUST be zero. 

valueTemp = 0 
```

```python
### Specify the number of iterations (days) for our recursive calculation. 

nDays = 500
```

```python
### The outer loop controls the recursive iterations. 

for day in range(nDays):
        
    ### The inner loop ranges over the action-space to calculate V(x,a) 
    ### for the current state. Here, action is the number of hours to rent 
    ### the point (range(6) is a list of the integers 0,1,2,3,4,5). 
    
    for action in range(6):
        
        ### The heart of the dynamic programming algorithm lies in this 
        ### single line. All other code supports this line by (1) creating 
        ### or importing the necessary objects; (2) controlling execution 
        ### of the line; (3) storing the output of this line; (4) displaying 
        ### the output of this line. 
        
        valueActionFunction[action] = 13000*(1-math.exp(-action/13)) \
        - 700*action + valueTemp*math.exp(-action/13)
    
    ### Update the current value function V(x) = max V(x,a). 
    
    valueTemp = max(valueActionFunction)
    
    ### Append the current value of V(x) to the value function object 
    ### valueFunction.
    
    valueFunction.append(valueTemp)
    
    ### Capture the current optimal action, then append it to the policy 
    ### function object policyFunction.   
    
    policyFunction.append(valueActionFunction.index(max(valueActionFunction)))
```

```python
### Plot the policy function against the number of remaining search days. 
    
plt.plot(range(1,nDays+1),policyFunction,linewidth=2) 
plt.xlabel('Days Remaining for Search') 
plt.ylabel('Optimal Number of Hours to Rent Boat') 
plt.title('Optimal Policy for Treasure Seeking SDP') 
plt.show() 
```

```python
### Plot the value function against the number of remaining search days. 
    
plt.plot(range(1,nDays+1),valueFunction,linewidth=2) 
plt.xlabel('Days Remaining for Search') 
plt.ylabel('Optimal Expected Reward') 
plt.title('Optimal Value Function for Treasure Seeking SDP') 
plt.show() 
```
