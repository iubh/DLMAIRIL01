## Functions

```python
###
### Functions  
###

### This function increments the state array. 

def incState( stateArray ):
    if ( stateArray.count(3) == len(stateArray) ):
        stateArray = [1]*len(stateArray)  
    else:
        ni = 0
        while stateArray[ni] == 3: ni=ni+1
        stateArray[ni] = stateArray[ni]+1
        if (ni > 0): 
            for i in range(ni): 
                stateArray[i]=1
  
    return stateArray  

### This function identifies the top disk of each peg for the state array input.

def topDisks ( stateArray ):
    n = len(stateArray)
    topDiskArray = array('i')
    topDiskArray = [n+1,n+1,n+1]

    for i in range(3): 
        if ( stateArray.count(i+1) > 0 ):
            ni = 0
            while (stateArray[ni] != (i+1)): ni = ni+1
            topDiskArray[i] = ni+1
                
    return topDiskArray

### This function converts a state array input to a state index.

def pHash ( stateArray ):  
    n = len(stateArray)
    sm = 0	    
    for i in range(n): 
        sm = sm + (3**i)*(stateArray[i]-1)  
    return sm

### This function creates a simple text graphic of a configuration.

def makeTowerPic ( stateArray ):  
    n = len(stateArray)
    towerPic = ['|','|','|']
    for i in range(n):
        towerPic[stateArray[n-1-i]-1]=towerPic[stateArray[n-1-i]-1]+str(n-i) 
   
    return towerPic
```

## Main program

```python
# Set the number of disks.

nDisk = 3
```

```python
# Create the value function. 

valueFunction = array('i') 
valueFunction = [0]*3**nDisk
```

```python
# Create temporary value function.

valueFunctionOld = array('i') 
valueFunctionOld = [0]*3**nDisk
```

```python
# Create policy function.

policyFunction = array('i')
policyFunction = [0]*3**nDisk
```

```python
# We need a large number that will exceed any admissible policy cost. 
# This can be the total number of states. 

bigM = 3**nDisk
```

```python
# Create value-action function V(x,a). We only required a fixed amount of memory. 

valueActionFunction = array('i')
valueActionFunction = [0]*6
```

```python
# Create array for admissible actions.

action = array('i')
action = [0]*6
```

```python
# Iterations continue while flagContinue equals True.
            
flagContinue = True

while flagContinue: 

    # Copy value function into temporary array.
        
    for iii in range(3**nDisk): valueFunctionOld[iii] = valueFunction[iii]
  
    # We will terminate iterations unless the value function changes.
    
    flagContinue = False
    
    # Create and initialize state array.
    
    stateArray = array('i')
    stateArray = [1]*nDisk
    
    # Create and initialize temporary state array.
    
    stateArrayTemp = array('i')
    stateArrayTemp = [1]*nDisk
    
    # Loop through all states, except for the terminal state, for which the 
    # value function is always zero. 
        
    for j in range(3**nDisk - 1):
      # Extract the top disks for the current array.
      topDiskArray = topDisks(stateArray)  
                
      # Initialize value-action function and action array.                 
      valueActionFunction = [bigM]*6
      action = [0]*6
       
      # We next consider the six possible moves. The IF statement checks 
      # for admissibility. If the action is not admissible, 
      # then V(x,a) = bigM, and that action will not be selected. 
        
      # Peg 1 to Peg 2.
      if (topDiskArray[0] < topDiskArray[1]):
        for iii in range(nDisk): stateArrayTemp[iii] = stateArray[iii]
        stateArrayTemp[topDiskArray[0]-1] = 2
        valueActionFunction[0] = 1 + valueFunctionOld[pHash(stateArrayTemp)]
        action[0] = pHash(stateArrayTemp)
                    
      # Peg 1 to Peg 3.                     
      if (topDiskArray[0] < topDiskArray[2]): 
        for iii in range(nDisk): stateArrayTemp[iii] = stateArray[iii]
        stateArrayTemp[topDiskArray[0]-1] = 3
        valueActionFunction[1] = 1 + valueFunctionOld[pHash(stateArrayTemp)]
        action[1] = pHash(stateArrayTemp)
     
      # Peg 2 to Peg 1.
      if (topDiskArray[1] < topDiskArray[0]):
        for iii in range(nDisk): stateArrayTemp[iii] = stateArray[iii]
        stateArrayTemp[topDiskArray[1]-1] = 1
        valueActionFunction[2] = 1 + valueFunctionOld[pHash(stateArrayTemp)]
        action[2] = pHash(stateArrayTemp)
    
      # Peg 2 to Peg 3.      
      if (topDiskArray[1] < topDiskArray[2]):
        for iii in range(nDisk): stateArrayTemp[iii] = stateArray[iii]
        stateArrayTemp[topDiskArray[1]-1] = 3
        valueActionFunction[3] = 1 + valueFunctionOld[pHash(stateArrayTemp)]
        action[3] = pHash(stateArrayTemp)
    
      # Peg 3 to Peg 1.
      if (topDiskArray[2] < topDiskArray[0]):
        for iii in range(nDisk): stateArrayTemp[iii] = stateArray[iii]
        stateArrayTemp[topDiskArray[2]-1] = 1
        valueActionFunction[4] = 1 + valueFunctionOld[pHash(stateArrayTemp)]
        action[4] = pHash(stateArrayTemp)
    
      # Peg 3 to Peg 2.
      if (topDiskArray[2] < topDiskArray[1]):
        for iii in range(nDisk): stateArrayTemp[iii] = stateArray[iii]
        stateArrayTemp[topDiskArray[2]-1] = 2
        valueActionFunction[5] = 1 + valueFunctionOld[pHash(stateArrayTemp)]
        action[5] = pHash(stateArrayTemp)
            
      # Bellmans Equation V(x) = min_a V(x,a) comes next.  
      valueFunction[j] = min(valueActionFunction)
            
      # Capture the optimal policy      
      policyFunction[j] = action[valueActionFunction.index(min(valueActionFunction))]
            
      # Increment the state array.       
      stateArray = incState(stateArray)         
    
      # If the value function has changed, set flagContinue = True so that 
      # the iterations will continue.
      if (valueFunction[j] != valueFunctionOld[j]): flagContinue = True
```

## Output solution

```python
# At this point, the value iteration algorithm has terminated.

# In order to display the solution, create a map from the state 
# index to the state array,  which is more easily interpreted. 
# This is stored in the matrix stateMap.
    
stateMap = numpy.zeros(shape=(3**nDisk,nDisk),dtype='i')
```

```python
# Re-initialize the state array.

stateArray = array('i')
stateArray = [1]*nDisk
```

```python
# Create the map by looping through each state.

for j in range(3**nDisk):
    stateMap[j] = stateArray  
    stateArray = incState(stateArray)            
```

```python
# Capture the optimal path from the policy function
# The number of transitions of the optimal path is given by
# valueFunction[0].

optimalPath = numpy.zeros(shape=(valueFunction[0]+1,nDisk),dtype='i')
a = 0
for j in range(valueFunction[0]+1): 
    optimalPath[j] = stateMap[a]
    a = policyFunction[a]
```

```python
# Print the optimal path

print('The optimal number of transitions is ',valueFunction[0],'.')
print('The optimal sequence of states is:')
print()

for i in range(valueFunction[0]+1): 
   currentTowerPic = makeTowerPic(optimalPath[i]) 
   print('---------------------------------   Stage ',i+1)
   for j in range(3):
       print(currentTowerPic[j])
  
print('---------------------------------   STOP')
```
